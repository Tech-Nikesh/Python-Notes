```python
###### Ratiobutton
from tkinter import *
root=Tk()
root.geometry("400x200")

def select():
    val=var.get()
    if val==1:
        print("Male Selected")
    elif val==2:
        print("Female Selector")

var=IntVar()
r1=Radiobutton(root,text="Male",variable=var,value=1,font=("",20,""),width=10,height=1)
r1.pack()

r2=Radiobutton(root,text="Female",variable=var,value=2,font=("",20,""),width=10,height=1)
r2.pack()

root.mainloop()
```

```python
#####  Cheakbutton
from tkinter import *
root=Tk()
root.geometry("400x200")

def select():
    val=var.get()
    val1=var1.get()

    if val==1 and val1==1:
        print("Both option selected")
    elif val==1 and val1==0:
        print("option 1 selected")
    elif val==0 and val1==1:
        print("option 2 selected")
    else:
        print("No option selected")

var=IntVar()
c1=Checkbutton(root,text="option 1",variable=var,command=select)
c1.pack()

var1=IntVar()
c2=Checkbutton(root,text="option 2",variable=var1,command=select)
c2.pack()

root.mainloop()
```

```python
#### Scale
from tkinter import *
root=Tk()
root.geometry("300x300")

def getvalue(n):
    print("Value=",n)

s=Scale(root,orient="vertical",command=getvalue).pack()

root.mainloop()
```

```python
#### Scroll
from tkinter import *
root=Tk()
root.geometry("300x300")

s=Scrollbar(root)
s.pack(side='right',fill=Y)

li=Listbox(root)
li.pack()


for x in range(100):
    li.insert(x,"option" +str(x+1))
s.config(command=li.yview)
 
root.mainloop()
```

```python
##### Menu

from tkinter import filedialog,messagebox
# root.Tk()
# root.geometry("400x300")

def new_f():
    new=Toplevel()
    new.geometry("400x400")
def open_f():
    f=filedialog.askopenfile()
    f.insert("0.0",f.read())

#filemenu
menubar=Menu(root)
filemenu=Menu(menubar,tearoff=0)

menubar.add_cascade(label="File",menu=filemenu,command=new_f)

filemenu.add_command(label="Save.File") 
filemenu.add_command(label="New.File") 
filemenu.add_command(label="Open.File") 
filemenu.add_command(label="Exit") 

#editmenu
menubar=Menu(root)
editmenu=Menu(menubar,tearoff=0)

menubar.add_cascade(label="Edit",menu=editmenu,command=new_f)

filemenu.add_command(label="copy") 
filemenu.add_command(label="cut") 
filemenu.add_command(label="paste") 
filemenu.add_command(label="find") 

#helpmenu
menubar=Menu(root)
helpmenu=Menu(menubar,tearoff=0)

menubar.add_cascade(label="help",menu=filemenu)

filemenu.add_command(label="A") 
filemenu.add_command(label="B") 
filemenu.add_command(label="C") 
filemenu.add_command(label="D") 


root.mainloop()
```
